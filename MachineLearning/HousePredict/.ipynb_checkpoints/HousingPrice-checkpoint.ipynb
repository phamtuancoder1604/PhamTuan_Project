```python
features = [
    "MasVnrArea",
    "BsmtFinSF1",
    "BsmtFinSF2",
    "BsmtUnfSF",
    "BsmtFullBath",
    "BsmtHalfBath",
    "Fireplaces",
    "WoodDeckSF",
    "OpenPorchSF",
    "EnclosedPorch",
    "3SsnPorch",
    "ScreenPorch",
    "Street",
    "RoofStyle",
    "RoofMatl",
    "Heating",
    "HeatingQC",
]
x = x.drop(features, axis = 1)
x
```

```python
categoricals = ['MSZoning',  'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle',   'Exterior1st',
       'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation', 
        'CentralAir', 'Electrical', 'KitchenQual', 'Functional',
       'PavedDrive', 'SaleType', 'SaleCondition']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_val, y_train, y_val = train_test_split(x,y,train_size = 0.8,test_size = 0.2, random_state = 0)
X_train.head()
```

```python
X_train.columns 
```

```python
selected_columns = [
    'MSSubClass', 'MSZoning', 'LotArea', 'LotShape', 'LandContour',
    'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1',
    'Condition2', 'BldgType', 'HouseStyle', 'OverallQual', 'OverallCond',
    'YearBuilt', 'YearRemodAdd', 'Exterior1st', 'Exterior2nd', 'ExterQual',
    'ExterCond', 'Foundation', 'TotalBsmtSF', 'CentralAir', 'Electrical',
    '1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'FullBath',
    'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual',
    'TotRmsAbvGrd', 'Functional', 'GarageCars', 'GarageArea', 'PavedDrive',
    'PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'SaleType', 'SaleCondition'
]

# Select only the specified columns
test = test[selected_columns]
test
```

```python
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OneHotEncoder , LabelEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer 
from sklearn.pipeline import Pipeline 
# numerical_cols = SimpleImputer(strategy = 'mean')
# categorical_cols = Pipeline(steps = [('imputer',SimpleImputer(strategy = 'most_frequent')),
#                                      ('onehot',OneHotEncoder(handle_unknown = 'ignore'))])
# preprocessor = ColumnTransformer(transformers = [('num',numerical_cols,numericals ),
#                                                 ('cat',categorical_cols, categoricals)])
```

```python
if isinstance(y_train, pd.DataFrame):
    print("df is a DataFrame")
else:
    print("df is NOT a DataFrame")
```

```python
od = OrdinalEncoder()
X_train[categoricals] = od.fit_transform(X_train[categoricals])
X_val[categoricals] = od.fit_transform(X_val[categoricals])
```

```python
X_train.columns
```

```python
test[categoricals] = od.fit_transform(test[categoricals])
test
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators = 100 , random_state = 0)
# pipeline = Pipeline(steps = [('preprocessor',preprocessor),
#                             ('model',model)])
model.fit(X_train,y_train)
predict = model.predict(X_val)
predict
```
